### Creating video from predicted frames

```python
import cv2
import numpy as np
from cv2 import VideoWriter, VideoWriter_fourcc
width = 640
height = 480
width2 = 64
height2 = 64
FPS = 5 #Saniyedeki frame sayısı azaltılarak saniye artırılabilir.
fourcc = VideoWriter_fourcc(*'mp4v')
name='/content/drive/MyDrive/predicted_videos/multiscalecircle_comparison.mp4'
name2='/content/drive/MyDrive/predicted_videos/multiscalecircle_predicted.mp4'
video = VideoWriter(name, fourcc, float(FPS), (width, height))
video2 = VideoWriter(name2, fourcc, float(FPS), (width2, height2))
for i in range(10):
  img = cv2.imread('/content/drive/MyDrive/multi_scale_GAN_circle_prediction/comparison{}.png'.format(i))
  video.write(img)
  img2 = cv2.imread('/content/drive/MyDrive/multi_scale_GAN_circle_prediction/NORM_OUTPUTS{}.png'.format(i))
  video2.write(img2)
video.release()
video2.release()
```
